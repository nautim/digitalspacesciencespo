```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the data
path = '/Users/nautim/Desktop/CEU DNDS/Winter 22:23/Diving into public space/data_project/' # use your path

terror231222 = pd.read_csv(path+'terror231222_datausers_cleaned.csv.gz', compression='gzip', dtype='str')
terror230421 = pd.read_csv(path+'terror230421_datausers_cleaned.csv.gz', compression='gzip', dtype='str')
terror291020 = pd.read_csv(path+'terror291020_datausers_cleaned.csv.gz', compression='gzip', dtype='str')
terror281019 = pd.read_csv(path+'terror281019_datausers_cleaned.csv.gz', compression='gzip', dtype='str')
```

```python
terror231222.created_at = pd.to_datetime(terror231222.created_at)
terror230421.created_at = pd.to_datetime(terror230421.created_at)
terror291020.created_at = pd.to_datetime(terror291020.created_at)
terror281019.created_at = pd.to_datetime(terror281019.created_at)
```

```python
# plot the number of tweets per day for the whole dataset
terror231222.created_at.dt.date.value_counts().sort_index().plot()
plt.xticks(ticks=terror231222.created_at.dt.date.value_counts().sort_index().index, 
           labels=terror231222.created_at.dt.date.value_counts().sort_index().index,
           rotation=45)
plt.axvline(x='2022-12-23', color='red', alpha=0.5, linestyle='--');
```
